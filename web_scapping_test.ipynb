```python
import requests  #send a http request to the server and return the response in the form of html content of that website from the server.
                              
#beautifulsoup--> used to work with html contents so that we can extract useful data from htlm content,xml also
from bs4 import BeautifulSoup

url = 'https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time-in-python'

response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

question_title = soup.find('a', class_='question-hyperlink').get_text() #find is used to  fetch single data present in a tag
question_body = soup.find('div', class_='s-prose js-post-body').get_text() #find_all is used fetch all data present in the given tag.

answers = soup.find_all('div', class_='answer')
print(f"Question: {question_title}")
print(f"Question Details: {question_body}\n")

for idx, answer in enumerate(answers):
    answer_body = answer.find('div', class_='s-prose js-post-body').get_text() #This will return the result without tags
    print(f"Answer {idx}: {answer_body}\n")
```
